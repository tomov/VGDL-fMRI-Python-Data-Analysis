```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

<H1>Calculate behavior RDM for each subject</H1>

```python
behavior_all = np.genfromtxt('behavior_similarity_all.csv', dtype=['S16','i8','S16','S16','i8','i8','i8','i8'], delimiter=',', names=True)
len(behavior_all)
```

```python
# Get an average similarity matrix for across all behavioral subjects
games = {'bait':0,'chase':1,'helper':2,'lemmings':3,'plaqueAttack':4,'zelda':5}
# This assumes each subject is asked 15 questions
num_q = 15
num_subjects = len(behavior_all) / num_q
behavior_RDM = np.zeros(shape=(6,6))
ans_freq = np.zeros(shape=(6,6))
for i in range(int(num_subjects)):
    for j in range(num_q):
        lineno = i*num_q+j
        game1 = behavior_all[lineno][2].decode('UTF-8')
        game2 = behavior_all[lineno][3].decode('UTF-8')
        if (game1 in games) and (game2 in games):
            game1_id = games[game1]
            game2_id = games[game2]
            dissim_rule = (1.0 - int(behavior_all[lineno][6])/10.0)
            behavior_RDM[game1_id][game2_id] += dissim_rule
            behavior_RDM[game2_id][game1_id] += dissim_rule
            ans_freq[game1_id][game2_id] += 1
            ans_freq[game2_id][game1_id] += 1
for i in range(len(games)):
    for j in range(i):
        behavior_RDM[i][j] /= ans_freq[i][j]
        behavior_RDM[j][i] /= ans_freq[i][j]
# TODO 
# 1. do I need to account for each subject's "average" score is different
# 2. The above code only recorded for "rule". The other three metrics can be done the same way

# Output, in the above games order
print(behavior_RDM)
np.save('behavior_RDM', behavior_RDM)
```

<H2> Test with a random group of voxel, from all subjects. Get the RDM for each, and compare each with behavior_RDM. Each subject should get a (rho,p)</H2>

```python
all_subjects_spearman_rho = []
all_subjects_spearman_p = []
for subject_id in range(bold_data_game.shape[2]):
    subject = bold_data_game[:,:,subject_id]
    region = subject[:,0:1000]
    print(region.shape)
    neural_RDM = 1-np.corrcoef(region)
    subject_spearman = scipy.stats.spearmanr(neural_RDM, behavior_RDM,axis=None)
    all_subjects_spearman_rho.append(subject_spearman.correlation)
    all_subjects_spearman_p.append(subject_spearman.pvalue)
print(all_subjects_spearman_rho, '\n', all_subjects_spearman_p)
```

<H3>t-test across subjects, for the same voxel region (only one region). Use 1-sample t-test</H3>
