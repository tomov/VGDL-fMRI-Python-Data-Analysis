```python
from nilearn import datasets
from nilearn import surface
from nilearn import plotting
from nilearn.input_data import NiftiMasker, NiftiLabelsMasker
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
path = '/Users/Daphne/data' # local directory

# these are the underlay for plotting (brain_nii) and the mask to make the volume (mask_nii)
mean_nii = nib.load(os.path.join(path, 'mean.nii')) # mean.nii is the anatomical image (sometimes called structural_nii)

# masks
mask_nii = nib.load(os.path.join(path, 'mask.nii')) # mask_nii is the functional mask, this selects the brain voxels
nosmooth_mask_nii = nib.load(os.path.join(path, 'mask_nosmooth.nii'))
```

# Looking into masks

> `.dataobj: ` Object containg image data. It should be some object that retuns an array from np.asanyarray. It should have a shape attribute or property


Below you see that the `mask.nii` file contains the whole brain mask. Here we overlay the mask on top of the anatomical image, `mean.nii`.

Next we want to
1. Create a volume from `mask.nii`
2. Put the correlations in this voxel space
3. Create a nifti image from this
4. Plot the statistical map


```python
brain_mask = np.array(mask_nii.dataobj) # get the mask from nii image
```

```python
brain_mask.shape
```

```python
np.where(brain_mask)
```

```python
masknii_nifti = nib.Nifti1Image(brain_mask, mask_nii.affine, mask_nii.header) 
#masknii_nifti = nib.Nifti1Image(brain_mask, nosmooth_mask_nii.affine, nosmooth_mask_nii.header) 
```

```python
# plot the data as statmap
f, ax = plt.subplots(1,1, figsize = (12, 5))
plotting.plot_stat_map(
    stat_map_img=masknii_nifti, 
    bg_img=mean_nii,
    threshold=0.2, 
    axes=ax,
    cut_coords=[-30, -4, 5]
);
```
